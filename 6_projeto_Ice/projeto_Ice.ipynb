```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# -----------------------------------------------------
# Ler um arquivo texto (csv)
# df === data frame
df = pd.read_csv('icecreamsales.csv')
df
```

```python
# Convers√£o de Fahrenheit para Celsius
celsius = (df['Temperature'] - 32) * 5 / 9
df['Temperature'] = celsius

df
```

```python
# Colocar temperatura em ordem decrescente
result = df.sort_values('Temperature', ascending = False)
result
```

```python
# Criar o gr√°fico com todos os dados, e salvar em pdf. 
plt.title('Ice cream Truck')

plt.ylabel('Sales (US$')
plt.xlabel('Temperatura (Celsius)')

plt.plot(result['Temperature'], result['Sales'], 'r.--', label = 'US$')

plt.legend()
plt.savefig('icecreamSales.pdf')
plt.show()
```
