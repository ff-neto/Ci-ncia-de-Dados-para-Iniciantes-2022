```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
# -----------------------------------------------------
# plot -> criar gráfico de linhas e comparar com três países
plt.title('Valor em combustivel em US$')

plt.plot(gas['Year'], gas['Australia'], 'r.--', label = 'Australia')
plt.plot(gas['Year'], gas['Italy'], 'y.--', label = 'Italy')
plt.plot(gas['Year'], gas['USA'], 'b.--', label = 'USA')
plt.plot(gas['Year'], gas['Japan'], 'g.--', label = 'Japan')

plt.xticks(gas['Year'][::2])

plt.xlabel('Years')
plt.ylabel('US$')

plt.legend()
plt.show()
```
